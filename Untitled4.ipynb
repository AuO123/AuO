```python
import jieba
from gensim.models import Word2Vec
import numpy as np

# 示例语料（可替换为你的真实文本）
corpus = [
    "这家餐厅的菜很好吃，味道很美味。",
    "我们周末经常去餐厅聚会，很热闹。",
    "安静的环境适合学习，不适合聚会。",
    "蟑螂是一种让人讨厌的昆虫。",
    "环境对人的心情影响很大。"
]

# 分词处理
sentences = []
for text in corpus:
    words = jieba.lcut(text)
    sentences.append(words)

print("分词后的语料：")
for s in sentences:
    print(s)
```

```python
# 训练 Skip-Gram 模型（sg=1 表示 Skip-Gram，sg=0 表示 CBOW）
model = Word2Vec(
    sentences=sentences,
    vector_size=100,  # 词向量维度
    window=5,         # 上下文窗口大小
    min_count=1,      # 忽略出现次数少于1的词
    workers=4,
    sg=1,             # 关键：使用 Skip-Gram
    epochs=10         # 训练轮数
)

print("Skip-Gram 模型训练完成！")
```

```python
# 获取“环境”的词向量
env_vec = model.wv["环境"]
print("“环境”的词向量：")
print(env_vec)
print("\n词向量形状：", env_vec.shape)
```

```python
# 查找与“好吃”最相似的前3个词
top3_similar = model.wv.most_similar("好吃", topn=3)
print("与“好吃”语义最接近的3个词：")
for word, score in top3_similar:
    print(f"{word}：相似度 {score:.4f}")
```

```python
# 计算余弦相似度
sim_good_tasty = model.wv.similarity("好吃", "美味")
sim_good_cockroach = model.wv.similarity("好吃", "蟑螂")

print(f"“好吃”和“美味”的相似度：{sim_good_tasty:.4f}")
print(f"“好吃”和“蟑螂”的相似度：{sim_good_cockroach:.4f}")
```

```python
# 向量运算：餐厅 + 聚会 - 安静
result = model.wv.most_similar(
    positive=["餐厅", "聚会"],
    negative=["安静"],
    topn=1
)

print("“餐厅+聚会-安静”的最相关结果：")
print(f"{result[0][0]}：相似度 {result[0][1]:.4f}")
```
